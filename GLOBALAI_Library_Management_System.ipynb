```python
class Library:
    def __init__(self, books_file):
        self.books = []
        self.books_file = books_file
        with open(self.books_file, 'a+') as f:
            for line in f:
                book_info = line.strip().split(',')
                book = Book(book_info[0], book_info[1], book_info[2], int(book_info[3]))
                self.books.append(book)

    def __del__(self):
        with open(self.books_file, 'a+') as f:
            f.close()

    def list_books(self):
      """
          This method will list all the books in the books.txt file.
          i. Read the contents of the file.
          ii. Add each line to a list using splitlines() method of the string object.
          iii. Now each element of the list holds information about a single book.
          Print book names and authors using this information.
      """


      with open(self.books_file, 'r') as f:
            for line in f:
                book_info = line.strip().split(',')
                print(f"{book_info[0]} by {book_info[1]}")

    def add_book(self, name, author, release_date, pages):
        """
          This method will add a book to a books.txt file.
          i. Ask user input for book title, book author, first release year and number
          of pages
          ii. Create a string with this information. Add book title then comma then
          author then comma etc.
          iii. Append this line to the file.
        """



        with open(self.books_file, 'a') as f:
            f.write(f"{name},{author},{release_date},{pages}\n")
            self.books.append(Book(name, author, release_date, pages))

    def remove_book(self, name):
        # This method will delete the book with the given title from the books.txt.
        # i. Ask the user input for book title.
        book_to_remove = input("Enter book name to remove: ")

        # ii. Read the file contents and add book to a list (just like you did while creating a list books method).
        with open(self.books_file, 'r') as f:
            self.books = []
            for line in f:
                book_info = line.strip().split(',')
                book = Book(book_info[0], book_info[1], book_info[2], int(book_info[3]))
                self.books.append(book)

        # iii. Find the index of the book to be deleted in the list.
        book_index = -1
        for i, book in enumerate(self.books):
            if book.name == book_to_remove:
                book_index = i
                break

        # iv. Remove the book from the list.
        if book_index != -1:
            del self.books[book_index]
        else:
            print(f"Book '{book_to_remove}' not found.")
            return

        # v. Remove the contents of the books.txt.
        with open(self.books_file, 'w') as f:
            pass

        # vi. Add all elements of the list to
        with open(self.books_file, 'a') as f:
            for book in self.books:
                f.write(f"{book.name},{book.author},{book.release_date},{book.pages}\n")

class Book:
    def __init__(self, name, author, release_date, pages):
        self.name = name
        self.author = author
        self.release_date = release_date
        self.pages = pages

    def __str__(self):
        return f"{self.name} by {self.author} ({self.release_date}), {self.pages} pages"

# Create an object named “lib” with “Library” class.
lib = Library("books.txt")

# Create a menu to interact with the “lib” object.
while True:
    print("*** MENU***")
    print("1) List Books")
    print("2) Add Book")
    print("3) Remove Book")
    print("4) Exit")

    # Get user input
    choice = input("Enter your choice: ")

    # Check user input
    if choice == "1":
        lib.list_books()
    elif choice == "2":
        name = input("Enter book name: ")
        author = input("Enter author name: ")
        release_date = input("Enter release date: ")
        pages = int(input("Enter number of pages: "))
        lib.add_book(name, author, release_date, pages)
    elif choice == "3":
        lib.remove_book(name)
    elif choice == "4":
        break
    else:
        print("Invalid choice")

# Close the file when the object is destroyed
del lib
```
